```python
# Problem 1: Salary Hike Based on Experience and Performance
%sql
SELECT 
    emp_name,
    department,
    salary,
    experience,
    performance_rating,

    CASE 
        WHEN experience >= 8 AND performance_rating = 'A' THEN salary * 1.20
        WHEN experience >= 5 AND performance_rating = 'B' THEN salary * 1.15
        WHEN performance_rating = 'C' THEN salary
        ELSE salary * 1.10
    END AS new_salary

FROM employee_data;
```

```python
# Problem 2: Bonus Calculation Based on Department and Performance
%sql
SELECT 
    emp_name,
    department,
    salary,
    performance_rating,

    CASE 
        -- For Finance department
        WHEN department = 'Finance' AND performance_rating = 'A' THEN salary * 0.20
        WHEN department = 'Finance' AND performance_rating = 'B' THEN salary * 0.15
        WHEN department = 'Finance' AND performance_rating = 'C' THEN salary * 0.05

        -- For Engineering department
        WHEN department = 'Engineering' AND performance_rating = 'A' THEN salary * 0.18
        WHEN department = 'Engineering' AND performance_rating = 'B' THEN salary * 0.12
        WHEN department = 'Engineering' AND performance_rating = 'C' THEN salary * 0.03

        -- Other departments
        ELSE salary * 0.10
    END AS bonus

FROM employee_data;
```

```python
# Problem 3: Categorizing Employees by Salary Range and Performance
%sql
SELECT 
    emp_name,
    department,
    salary,
    performance_rating,

    CASE 
        WHEN salary > 80000 AND performance_rating = 'A' THEN 'High Performer'
        WHEN salary BETWEEN 50000 AND 80000 AND performance_rating = 'B' THEN 'Mid Performer'
        WHEN salary < 50000 OR performance_rating = 'C' THEN 'Low Performer'
    END AS category

FROM employee_data;
```

```python
# Problem 4: Risk Assessment Based on Experience and Department
%sql
SELECT 
    emp_name,
    department,
    experience,

    CASE 
        -- HR Department
        WHEN department = 'HR' AND experience < 5 THEN 'High Risk'
        WHEN department = 'HR' AND experience >= 5 THEN 'Low Risk'

        -- Engineering & Finance Departments
        WHEN department IN ('Engineering', 'Finance') AND experience > 8 THEN 'Low Risk'
        WHEN department IN ('Engineering', 'Finance') AND experience <= 8 THEN 'Medium Risk'

        -- Other Departments
        ELSE 'Medium Risk'
    END AS risk_level

FROM employee_data;
```

```python
#Problem 1: Nested CASE for Performance and Salary Hike Based on Multiple Criteria
%sql
SELECT 
    emp_name,
    salary,
    experience,
    performance_rating,

    CASE 
        WHEN performance_rating = 'A' THEN
            CASE 
                WHEN salary > 80000 AND experience > 5 THEN salary * 1.25
                WHEN salary BETWEEN 50000 AND 80000 THEN salary * 1.15
                ELSE salary * 1.20
            END

        WHEN performance_rating = 'B' THEN
            CASE 
                WHEN experience > 5 THEN salary * 1.12
                ELSE salary * 1.10
            END

        WHEN performance_rating = 'C' THEN salary

    END AS new_salary

FROM employee_data;
```

```python
#problem 2:Department and Performance
%sql
SELECT 
    emp_name,
    department,
    salary,
    experience,
    performance_rating,

    CASE 
        WHEN department = 'Finance' THEN
            CASE 
                WHEN performance_rating = 'A' AND experience > 10 THEN salary * 0.25
                ELSE salary * 0.20
            END

        WHEN department = 'HR' THEN
            CASE 
                WHEN performance_rating = 'B' OR experience > 5 THEN salary * 0.15
                ELSE salary * 0.10
            END

        ELSE salary * 0.08
    END AS bonus

FROM employee_data;
```

```python
#problem 3:Nested CASE for Employee Categorization Based on Salary, Performance, and ExperienceScenario: Categorize employees based on their salary, performance rating, and experience:
%sql
SELECT 
    emp_name,
    salary,
    experience,
    performance_rating,

    CASE 
        WHEN salary > 70000 THEN
            CASE 
                WHEN performance_rating = 'A' AND experience > 8 THEN 'Top Performer'
                WHEN experience < 8 THEN 'Mid Performer'
                ELSE 'Mid Performer'
            END

        WHEN salary BETWEEN 50000 AND 70000 THEN
            CASE 
                WHEN performance_rating = 'A' THEN 'Rising Star'
                ELSE 'Average Performer'
            END

        ELSE 'Low Performer'
    END AS category

FROM employee_data;
```

```python
#4.Nested CASE for Tax Bracket Based on Salary and Experience
%sql
SELECT 
    emp_name,
    salary,
    experience,

    CASE 
        WHEN salary > 90000 THEN
            CASE 
                WHEN experience > 10 THEN '35%'
                ELSE '30%'
            END

        WHEN salary BETWEEN 60000 AND 90000 THEN
            CASE 
                WHEN experience > 5 THEN '25%'
                ELSE '20%'
            END

        ELSE '15%'
    END AS tax_rate

FROM employee_data;
```

```python
#Problem 5: Nested CASE for Promotion Eligibility Based on Performance, Salary, and Experience
%sql
SELECT 
    emp_name,
    salary,
    experience,
    performance_rating,

    CASE 
        WHEN performance_rating = 'A' THEN
            CASE 
                WHEN salary > 75000 AND experience > 7 THEN 'Eligible for Senior Role'
                ELSE 'Eligible for Junior Role'
            END

        WHEN performance_rating = 'B' THEN
            CASE 
                WHEN experience > 5 THEN 'Eligible for Consideration'
                ELSE 'Not Eligible'
            END

        ELSE 'Not Eligible'
    END AS promotion_status

FROM employee_data;
```
