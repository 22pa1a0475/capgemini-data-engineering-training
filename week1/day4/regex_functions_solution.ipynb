```sql
select * from regex_practice;
```

Numbers at beginning

```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '^[0-9]+', 0) AS start_numbers
FROM regex_practice;
```

Numbers at end

```sql
SELECT id, REGEXP_EXTRACT(mixed_value, '[0-9]+$', 0) AS q2_end_numbers
FROM regex_practice;
```

the first single character of the value

```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '^.', 0) AS first_char
FROM regex_practice;
     
```

the last single character 

```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '.$', 0) AS last_char
FROM regex_practice;
     
```

two consecutive numeric characters 

```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '\b[0-9]{2}\b', 0) AS two_digits
FROM regex_practice;
     
```

one numeric character that appears anywhere in the
value.

```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '[0-9]', 0) AS one_digit
FROM regex_practice;
     
```

country code present at the beginning of the phone number

```sql
SELECT phone,
REGEXP_EXTRACT(phone, '^\+?([0-9]+)', 1) AS country_code
FROM regex_practice;
```

the numeric portion that is present between alphabetic
characters.


```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '[a-zA-Z]+([0-9]+)[a-zA-Z]+', 1) AS middle_numbers
FROM regex_practice;
     
```

the text that appears before the at-symbol

```sql
SELECT email,
REGEXP_EXTRACT(email, '^[^@]+', 0) AS username
FROM regex_practice;
```

the text that appears after the at-symbol including the domain
name.


```sql
SELECT email,
REGEXP_EXTRACT(email, '@(.+)', 1) AS domain_full
FROM regex_practice;
     
```

only the domain name without including the at-symbol.

```sql
SELECT email,
REGEXP_EXTRACT(email, '@([^\.]+)', 1) AS domain_name
FROM regex_practice;
     
```

only the text that appears after the last dot in the email address.


```sql
SELECT email,
REGEXP_EXTRACT(email, '\.([^.]+)$', 1) AS extension
FROM regex_practice;
     
```

only alphabetic characters that appear together as a
continuous sequence



```sql

SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '[a-zA-Z]+', 0) AS alpha_seq
FROM regex_practice;
```

only numeric characters that appear together as a continuous
sequence.


```sql
SELECT mixed_value,
REGEXP_EXTRACT(mixed_value, '[0-9]+', 0) AS num_seq
FROM regex_practice;
```

the first three characters of the value

```sql
SELECT full_text,
SUBSTRING(full_text, 1, 3) AS first_three
FROM regex_practice;
     
```

the last two characters of the value

```sql

SELECT full_text,
RIGHT(full_text, 2) AS last_two
FROM regex_practice;
```

Employee number (between prefix & underscore)

```sql
SELECT full_text,
REGEXP_EXTRACT(full_text, '^[A-Z]+([0-9]+)_', 1) AS emp_number
FROM regex_practice;
     
```

Country code at end

```sql
SELECT full_text,
REGEXP_EXTRACT(full_text, '([0-9]+)$', 1) AS country_code
FROM regex_practice;
```

Text between underscores

```sql
SELECT full_text,
REGEXP_EXTRACT(full_text, '_(.*?)_', 1) AS middle_text
FROM regex_practice;
     
```

Country code after "+"

```sql
SELECT phone,
REGEXP_EXTRACT(phone, '^\+([0-9]+)', 1) AS plus_country_code
FROM regex_practice;
     
```
