LEVEL 1 (BASIC)

```sql

-- Show all employees whose salary is NULL
select * from employees_data4 where salary is null;
```

```sql
-- Show all orders where discount is NOT NULL
select * from orders_data4 where discount is not null;
```

```sql
-- Get products where category is NULL
select * from products_data4 where category is null;
```

```sql
-- Count number of employees with NULL manager_id
select count(*) as no_of_employees from employees_data4 where manager_id is null;
```

LEVEL 2 (ISNULL)


```sql
-- Replace NULL salary with 0
update employees_data4 set  salary=0 where salary is null;
select salary from employees_data4;

```

```sql
-- Replace NULL bonus with 1000
update employees_data4 set bonus=10000 where bonus is null;
select bonus from employees_data4;
```

```sql
-- Show order amount, if NULL replace with 500
update orders_data4 set amount=500 where amount is null;
select amount from orders_data4
```

```sql
-- Replace NULL stock with 0
update  products_data4 set stock=0 where stock is null;
select stock from products_data4;
```

LEVEL 3 (COALESCE)

```sql
-- Show employee earnings using:
--  salary, if NULL use bonus
select name,coalesce(salary,bonus) as earnings from employees_data4


-- Get customer payment:
--  amount → discount → 0
```

```sql
-- Show first available value:
--  salary → bonus → 0
select name,coalesce(salary,bonus,0) as earnings from employees_data4
```

```sql
-- Show product price:
--  price → 1000 (default)
select product_id ,coalesce(price,1000) as price from products_data4;
```

```sql
-- Get customer payment:
--  amount → discount → 0
select customer_name,coalesce(amount,discount,0) as payment from orders_data4

```

LEVEL 4 (NULLIF)

```sql

-- Convert salary to NULL if salary = 0
SELECT emp_id, name, NULLIF(salary, 0) AS salary
FROM employees_data4;
```

```sql
-- Convert discount to NULL if discount = 0
select customer_name ,nullif(discount,0) as discount from orders_data4
```

```sql
-- Use NULLIF to avoid divide by zero:
-- SELECT amount / NULLIF(discount, 0)
-- FROM Orders;

SELECT order_id,
       amount / NULLIF(discount, 0) AS result
FROM orders_data4;
```

```sql
-- Replace coupon_code with NULL if it is 'DISC10'
SELECT order_id,
       NULLIF(coupon_code, 'DISC10') AS coupon_code
FROM orders_data4;
```

```sql
-- Replace coupon_code with NULL if it is 'DISC10'
SELECT order_id,
       NULLIF(coupon_code, 'DISC10') AS coupon_code
FROM orders_data4;
```

LEVEL 5 (REAL-TIME SCENARIOS)


```sql
-- Calculate total earnings:
--    salary + bonus (handle NULL properly)
SELECT name,
       COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM employees_data4;
```

```sql
-- Show employees where:
--    both salary AND bonus are NULL
select * from employees_data4 where salary is null and bonus is null;
```

```sql
-- Show products where:
--    price is NULL but category is NOT NULL
select * from products_data4 where price is null and category is not null
```

```sql
-- Show orders where:
--    both amount and discount are NULL
select * from orders_data4 where amount is null and discount is null
```

LEVEL 6 (ADVANCED)

```sql
-- Show employee income:
--   COALESCE(salary, bonus, 1000)
select name,coalesce (salary,bonus,1000) as employee_income from employees_data4

```

```sql
-- Replace empty discount with NULL using NULLIF:
-- NULLIF(discount, 0)
select customer_name,nullif(discount,0) as discount from orders_data4
```

```sql
-- Show final payable amount:
--   amount - discount (handle NULL)
select order_id,amount-nullif(discount,0) as payable_amount from orders_data4
```
