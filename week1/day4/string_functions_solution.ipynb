```sql
-- LENGTH / CHAR_LENGTH
SELECT full_name, LENGTH(full_name), CHAR_LENGTH(full_name) FROM employees_data5
;
SELECT full_name, CHAR_LENGTH(full_name) FROM employees_data5;
```

```sql
-- 2. UPPER / LOWER
SELECT UPPER(department), LOWER(city) FROM employees_data5;
```

```sql
-- 3.TRIM / LTRIM / RTRIM
SELECT
TRIM(full_name),
LTRIM(full_name),
RTRIM(full_name)
FROM employees_data5;

```

```sql
-- 4. CONCAT
SELECT CONCAT_WS(' | ', emp_id, full_name, city)
FROM employees_data5;
```

```sql
-- 6. SUBSTRING / SUBSTR / MID
-- Extract part of string
SELECT SUBSTRING(email, 1, 7) FROM employees_data5;
SELECT SUBSTR(email, 1, 5) FROM employees_data5;
```

```sql
-- 7. LEFT / RIGHT
-- Extract characters from left or right
SELECT LEFT(full_name, 4), RIGHT(city, 3)
FROM employees_data5;
```

```sql
-- 8. INSTR
-- Position of substring
SELECT email, INSTR(email, '@') FROM employees_data5;
```

```sql
-- 9. LOCATE
-- Same as INSTR but with more flexibility
SELECT email, LOCATE('.', email) FROM employees_data5;
```

```sql
-- 10. REPLACE
-- Replace substring
SELECT REPLACE(department, 'Data', 'Big Data')
FROM employees_data5;
```

```sql
-- 11. REVERSE
-- Reverse a string
SELECT full_name, REVERSE(full_name)
FROM employees_data5;
```

```sql
-- 12. LPAD / RPAD
-- Pad characters on left or right
SELECT LPAD(emp_id, 5, '0') FROM employees_data5;
SELECT RPAD(city, 15, '*') FROM employees_data5;
```

```sql
-- 13. TRIM + REPLACE (Combined Usage)
SELECT TRIM(REPLACE(city, ' ', '')) FROM employees_data5;
```

```sql
-- 14. IFNULL
-- Handle NULL strings
SELECT full_name, IFNULL(remarks, 'No remarks')
FROM employees_data5;
```

```sql
-- 15. COALESCE
-- Return first non-NULL value
SELECT full_name, COALESCE(remarks, 'N/A')
FROM employees_data5;
```

```sql
-- 16. FIND_IN_SET
-- Search value in a comma-separated list (CSV)
SELECT FIND_IN_SET('Analytics', 'Data,Analytics,AI');
```
