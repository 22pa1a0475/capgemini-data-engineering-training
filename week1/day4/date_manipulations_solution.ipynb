```sql
CREATE TABLE orders_data (
order_id INT PRIMARY KEY,
customer_name VARCHAR(50),
order_date DATE,
order_timestamp TIMESTAMP,
delivery_date DATE,
order_amount DECIMAL(10,2)
);

INSERT INTO orders_data VALUES
(1, 'Karthik', '2024-01-15', '2024-01-15 10:30:45', '2024-01-20', 2500.00),
(2, 'Veena', '2024-02-18', '2024-02-18 18:45:20', '2024-02-22', 3200.50),
(3, 'Ravi', '2024-03-02', '2024-03-02 09:15:10', '2024-03-08', 4100.75),
(4, 'Anil', '2024-03-09', '2024-03-09 14:05:55', '2024-03-15', 1800.00),
(5, 'Suresh', '2024-01-07', '2024-01-07 23:55:00', '2024-01-12', 2900.00);
```

```sql
--  EXTRACTING YEAR, MONTH, DAY
SELECT
YEAR(order_date),
MONTH(order_date),
DAY(order_date)
FROM orders_data;
```

```sql
-- EXTRACT USING EXTRACT()
SELECT
EXTRACT(YEAR FROM order_date),
EXTRACT(MONTH FROM order_date),
EXTRACT(DAY FROM order_date)
FROM orders_data;
```

```sql
-- . MONTH NAME AND DAY NAME
SELECT
MONTHNAME(order_date),
DAYNAME(order_date)
FROM orders_data;
```

```sql
--  WEEKDAY AND DAY OF WEEK
SELECT
WEEKDAY(order_date),
DAYOFWEEK(order_date)
FROM orders_data;
```

```sql
-- IDENTIFY WEEKENDS (SATURDAY & SUNDAY)
-- Method 1: Using DAYNAME
SELECT order_id, order_date
FROM orders_data
WHERE DAYNAME(order_date) IN ('Saturday', 'Sunday');
-- Method 2: Using DAYOFWEEK
SELECT order_id, order_date
FROM orders_data
WHERE DAYOFWEEK(order_date) IN (1, 7);
```

```sql
-- . DATE ARITHMETIC
SELECT order_date, DATE_ADD(order_date,  5)
FROM orders_data;
```

```sql
-- . DATE ARITHMETIC
--  Subtract Days
SELECT order_date, DATE_SUB(order_date, 3)
FROM orders_data;
```

```sql
-- Add Months
SELECT ADD_MONTHS(order_date,  1 ) FROM orders_data;
```

```sql
-- subtract Months
SELECT ADD_MONTHS(order_date,  -1) FROM orders_data;

```

```sql
-- Add  Years
SELECT ADD_MONTHS(order_date, -12) AS minus_one_year
FROM orders_data;
```

```sql
-- sub  Years
SELECT ADD_MONTHS(order_date, -36) AS minus_three_years
FROM orders_data;
```

```sql
--  DATEDIFF (Difference in Days)
SELECT
order_id,
DATEDIFF(delivery_date, order_date) AS delivery_days
FROM orders_data;
```

```sql
--  TIMESTAMPDIFF (Difference in Units)
SELECT
TIMESTAMPDIFF(DAY, order_date, delivery_date) AS days_diff,
TIMESTAMPDIFF(MONTH, order_date, delivery_date) AS months_diff
FROM orders_data;
```

```sql
-- 16. DATE FORMAT (IMPORTANT)
SET spark.sql.legacy.timeParserPolicy = LEGACY;
SELECT DATE_FORMAT(order_date, '%d-%m-%Y') FROM orders_data;

```

```sql
SELECT DATE_FORMAT(order_date, '%M %d, %Y') FROM orders_data;
```

```sql
--  STRING TO DATE (STR_TO_DATE)
-- Convert string to DATE
SELECT TO_DATE('15-01-2024', 'dd-MM-yyyy') AS parsed_date;


```

```sql
-- Convert string to TIMESTAMP
SELECT TO_TIMESTAMP('15-01-2024', 'dd-MM-yyyy') AS parsed_timestamp;
```

```sql
-- FINANCIAL YEAR LOGIC (CASE)
SELECT order_date,
CASE
WHEN MONTH(order_date) >= 4 THEN CONCAT(YEAR(order_date), '-', YEAR(order_date)+1)
ELSE CONCAT(YEAR(order_date)-1, '-', YEAR(order_date))
END AS financial_year
FROM orders_data;
```

```sql
-- TIMESTAMP FORMATTING
-- Format timestamp as dd-MM-yyyy HH:mm:ss
SELECT DATE_FORMAT(order_timestamp, 'dd-MM-yyyy HH:mm:ss') AS formatted_timestamp
FROM orders_data;
```

```sql
--  FILTER BY MONTH (JAN, FEB, MAR)
SELECT *
FROM orders_data
WHERE MONTH(order_date) = 1; -- January
```

```sql
-- FINANCIAL YEAR LOGIC (CASE)
SELECT order_date,
CASE
WHEN MONTH(order_date) >= 4 THEN CONCAT(YEAR(order_date), '-', YEAR(order_date)+1)
ELSE CONCAT(YEAR(order_date)-1, '-', YEAR(order_date))
END AS financial_year
FROM orders_data;
```

```sql
-- Orders in Last 7 Days
SELECT *
FROM orders_data
WHERE order_date >= CURDATE() - INTERVAL 7 DAY;
```

```sql
-- Orders Placed Today
SELECT *
FROM orders_data
WHERE DATE(order_timestamp) = CURDATE();
```
