group by with conditions

```sql
--6.total salary for departments where the total salary exceeds 100,000
select department,sum(salary) as total_salary from employee group by department having total_salary>100000
```

```sql
--7.Count the number of employees in departments with more than 2 employees.
select department,count(*) as no_of_employees from employee group by department having no_of_employees>2
```

```sql
--8.the average salary for employees who joined after 2021-01-01, grouped by department.
select department,avg(salary) as avg_salary from employee where joining_date>'2021-01-01' group by department
```

```sql
--9.Find the departments where the maximum salary is greater than 75,000.
select department ,max(salary) as max_salary from employee group by department having max_salary>75000
```

```sql
--10.List the departments where the total salary is less than 150,000.
select department ,sum(salary) as total_salary from employee group by department having total_salary>150000
```

```sql
--11.the total number of employees grouped by department, but only include departments with more than 1 employee.
select department,count(*) as no_of_employees from employee group by department having  no_of_employees>1
```

```sql
--12.Calculate the total salary of each department and show only those where the total exceeds 125,000.
select department,sum(salary) as total_salary from employee group by department having total_salary>125000
```

```sql
--13.Count the number of employees in each department, but include only departments with more than 2 employees.
select department,count(*) as no_of_employees from employee group by department having no_of_employees>2
```

```sql
-- 14.Find the average salary in each department where the average salary is above 60,000.
SELECT department, AVG(salary) AS avg_salary
FROM employee
GROUP BY department
HAVING AVG(salary) > 60000;
```

```sql
-- 15.Show departments where the sum of salaries is between 100,000 and 200,000.
SELECT department, SUM(salary) AS total_salary
FROM employee
GROUP BY department
HAVING SUM(salary) BETWEEN 100000 AND 200000;
```

GROUP BY with JOINs

```sql
-- 16. Find the total sales amount for each employee.
SELECT e.emp_name, SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
-- 17. List the number of sales made by each employee.
SELECT e.emp_name, COUNT(*) AS sales_count
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
-- 18. Find the total sales amount grouped by product.
SELECT product, SUM(amount) AS total_sales
FROM sales
GROUP BY product;
```

```sql
-- 19. Calculate the average sales amount grouped by product
SELECT product, AVG(amount) AS avg_sales
FROM sales
GROUP BY product;
```

```sql
-- 20. Find employees who have made more than 2 sales, grouped by their names.
SELECT e.emp_name, COUNT(*) AS sales_count
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name
HAVING COUNT(*) > 2;
```

Advanced GROUP BY

```sql
-- 21. Calculate the total salary and total sales amount for each employee.
SELECT e.emp_name,
       SUM(e.salary) AS total_salary,
       SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
-- 22. Count the number of unique products sold by each employee.
SELECT e.emp_name, COUNT(DISTINCT s.product) AS unique_products
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
-- 23. Find the highest sales amount made by each employee.
SELECT e.emp_name, MAX(s.amount) AS max_sale
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

```sql
-- 24. Calculate the total sales amount grouped by product and filtered by products where the total
-- exceeds 50,000
SELECT e.department, AVG(s.amount) AS avg_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY avg_sales DESC
LIMIT 1;
```

```sql
-- 25. Find the departments with the highest average sales amount.
SELECT e.department, AVG(s.amount) AS avg_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY avg_sales DESC
LIMIT 1;
```

Real-World Scenarios

```sql
-- 26. Find the department with the highest total sales amount.
SELECT e.department, SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY total_sales DESC
LIMIT 1;
```

```sql
-- 27. Show the top 3 employees with the highest total sales amount, grouped by employee names.
SELECT e.emp_name, SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name
ORDER BY total_sales DESC
LIMIT 3;
```

```sql
-- 28. Calculate the total number of employees and the average salary, grouped by the year of joining.
SELECT e.department, SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.department;
```

```sql
-- 29. Find the total sales amount for each department (using a join between Employee and Sales).
SELECT e.department, SUM(s.amount) AS total_sales
FROM employee e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.department;
```

```sql
-- 30. Show employees who have not made any sales, grouped by their department
SELECT e.department, e.emp_name
FROM employee e
LEFT JOIN sales s ON e.emp_id = s.emp_id
WHERE s.emp_id IS NULL;
```
