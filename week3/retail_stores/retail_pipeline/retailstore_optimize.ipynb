```python
%sql
-- STEP 1: USE CATALOG & SCHEMA

USE CATALOG retail_store_sales;
USE SCHEMA gold;


-- STEP 2: CREATE / REFRESH PHYSICAL TABLES
-- (Convert DLT views → Delta tables)

CREATE OR REPLACE TABLE fact_sales_physical AS
SELECT * FROM fact_sales;

CREATE OR REPLACE TABLE daily_store_sales_physical AS
SELECT * FROM daily_store_sales;

CREATE OR REPLACE TABLE discount_impact_physical AS
SELECT * FROM discount_impact_analysis;

CREATE OR REPLACE TABLE top_products_physical AS
SELECT * FROM top_products;


-- STEP 3: OPTIMIZE (FILE COMPACTION)

OPTIMIZE fact_sales_physical;
OPTIMIZE daily_store_sales_physical;
OPTIMIZE discount_impact_physical;
OPTIMIZE top_products_physical;


-- STEP 4: ZORDER (QUERY PERFORMANCE)

-- Fact table → most important
OPTIMIZE fact_sales_physical
ZORDER BY (customer_sk, product_sk, store_sk);

-- Time-based queries
OPTIMIZE daily_store_sales_physical
ZORDER BY (order_date);

OPTIMIZE discount_impact_physical
ZORDER BY (order_date);


-- STEP 5: OPTIONAL (ADVANCED)
-- CLEAN OLD FILES (Mention in interview)

VACUUM fact_sales_physical RETAIN 168 HOURS;
```
