Risk Score

```python
from pyspark.sql.functions import col, sum

# Step 1: Create clean base dataframe (IMPORTANT FIX)
df_clean = customers.alias("c") \
    .join(valid_policies.alias("p"), col("c.customer_id") == col("p.customer_id")) \
    .join(valid_claims.alias("cl"), col("p.policy_id") == col("cl.policy_id"), "left") \
    .select(
        col("c.customer_id").alias("customer_id"),
        col("c.name").alias("name"),
        col("c.city").alias("city"),
        col("p.policy_id").alias("policy_id"),
        col("p.premium").alias("premium"),
        col("cl.claim_amount").alias("claim_amount")
    )

# Step 2: Total Premium
premium_df = df_clean.groupBy("customer_id", "name", "city") \
    .agg(sum("premium").alias("total_premium"))

# Step 3: Total Claim
claim_df = df_clean.groupBy("customer_id") \
    .agg(sum("claim_amount").alias("total_claim"))

# Step 4: Join safely
customer_metrics = premium_df.join(claim_df, on="customer_id", how="left")

# Step 5: Handle nulls
customer_metrics = customer_metrics.fillna(0, ["total_claim"])

# Step 6: Risk score
customer_metrics = customer_metrics.withColumn(
    "risk_score",
    col("total_claim") / col("total_premium")
)

# Step 7: Output
customer_metrics.display()   # or .show()
```

City-wise Distribution

```python
city_metrics = customer_metrics.groupBy("city") \
    .agg(
        sum("total_premium").alias("city_premium"),
        sum("total_claim").alias("city_claim")
    )
```

Create Temp Views

```python
customer_metrics.createOrReplaceTempView("customer_metrics")
valid_claims.createOrReplaceTempView("claims")
```

Top 3 Risky Customers per City

```python
%sql
WITH ranked_customers AS (
    SELECT 
        customer_id,
        name,
        city,
        total_premium,
        total_claim,
        risk_score,
        RANK() OVER (PARTITION BY city ORDER BY risk_score DESC) AS rnk
    FROM customer_metrics
)
SELECT *
FROM ranked_customers
WHERE rnk <= 3;
```

Customers with Increasing Claims (Month-over-Month)

```python
%sql
WITH monthly_claims AS (
    SELECT 
        policy_id,
        DATE_FORMAT(claim_date, 'yyyy-MM') AS month,
        SUM(claim_amount) AS total_claim
    FROM claims
    GROUP BY policy_id, month
),
lagged_claims AS (
    SELECT 
        *,
        LAG(total_claim) OVER (PARTITION BY policy_id ORDER BY month) AS prev_claim
    FROM monthly_claims
)
SELECT *
FROM lagged_claims
WHERE total_claim > prev_claim;
```

Rank Customers Based on Risk Score (City-wise)

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank, row_number, col

# Window partitioned by city
window_spec = Window.partitionBy("city").orderBy(col("risk_score").desc())

customer_ranked = customer_metrics.withColumn(
    "rank", rank().over(window_spec)
).withColumn(
    "dense_rank", dense_rank().over(window_spec)
).withColumn(
    "row_number", row_number().over(window_spec)
)

customer_ranked.display()
```

2. Rank Agents Based on Total Premium

```python
from pyspark.sql.functions import col, sum
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank, row_number

# Step 1: Clean join with aliases (FIX ambiguity)
agent_df = valid_policies.alias("p") \
    .join(policy_agent.alias("pa"), col("p.policy_id") == col("pa.policy_id")) \
    .join(agents.alias("a"), col("pa.agent_id") == col("a.agent_id")) \
    .select(
        col("a.agent_id").alias("agent_id"),
        col("a.name").alias("agent_name"),
        col("p.premium").alias("premium")
    )

# Step 2: Aggregate premium
agent_perf = agent_df.groupBy("agent_id", "agent_name") \
    .agg(sum("premium").alias("total_premium"))

# Step 3: Window (FIX warning by adding partition if needed)
window_agent = Window.orderBy(col("total_premium").desc())

# Step 4: Ranking
agent_ranked = agent_perf.withColumn(
    "rank", rank().over(window_agent)
).withColumn(
    "dense_rank", dense_rank().over(window_agent)
).withColumn(
    "row_number", row_number().over(window_agent)
)

# Step 5: Display
agent_ranked.display()
```
